```python
# 导入数据分析包 pandas
import pandas as pd
# 导入类型库
import types
# 导入复制库
import copy
# 定义数据所在路径
TRAIN_PATH = '../dataset/voice.csv'
TEST_PATH = '../dataset/voice.csv'

# 利用pandas读入csv数据
train_data = pd.read_csv(TRAIN_PATH)
test_data = pd.read_csv(TEST_PATH)
```

```python
# 特征索引列表
features_index = []
# 标签索引列表
label_index = []
for feature in train_data.columns:
    features_index.append(feature)
label_index .append(features_index.pop())
print("特征索引")
print(features_index)
print("标签索引")
print(label_index)
```

```python
# 储存对应类型特征索引列表的字典
features_index_corespond_dtype_dict = dict()
# 特征类型的集合
features_dtype_set = []
# 特征类型的列表
features_dtype_list = []
for feature in features_index:
    features_dtype_list.append(train_data[feature].dtype)
features_dtype_set = set(features_dtype_list)
for dtype in features_dtype_set:
    new_list_for_features_index_corespond_dtype_dict = []
    features_index_corespond_dtype_dict[str(dtype)] = new_list_for_features_index_corespond_dtype_dict
for feature in features_index:
    features_index_corespond_dtype_dict[str(
        train_data[feature].dtype)].append(feature)
for feature_index, list in features_index_corespond_dtype_dict.items():
    print("类型为{}的特征索引".format(feature_index))
    print(list)
```

```python
# 数字类型的特征索引
numeric_features_index = []
# 整数的特征索引
int_features_index = []
# 浮点数的特征索引
float_features_index = []
for feature in features_index:
    if(train_data[feature].dtype == int):
        int_features_index.append(feature)
    elif(train_data[feature].dtype == float):
        float_features_index.append(feature)
numeric_features_index.extend(copy.deepcopy(int_features_index))
numeric_features_index.extend(copy.deepcopy(float_features_index))

```

```python
# 离散类型的特征索引列表
class_features_threshold = 20
class_features = []
nunique_of_features = train_data.nunique()
for feature in features_index:
    if(nunique_of_features[feature] < class_features_threshold):
        class_features.append(feature)
print('离散型特征索引')
print(class_features)
print("特征对应不同值的个数")
print(nunique_of_features)
```
